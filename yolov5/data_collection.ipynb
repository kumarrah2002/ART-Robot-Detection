```python
import cv2
import os
```

```python
directory = 'S:/selfdrivingcar/data'
if not os.path.exists(directory):
    os.makedirs(directory)
```

```python
def create_frames(cam, photo_id):
    currentframe = 0
    while(True):
        # reading from frame
        ret,frame = cam.read()

        if ret:
            # if video is still left continue creating images
            name = 'S:/selfdrivingcar/data/frame_' + str(currentframe + photo_id) + '.jpg'
            print ('Creating...' + name)

            # writing the extracted images
            cv2.imwrite(name, frame)

            # increasing counter so that it will
            # show how many frames are created
            currentframe += 1
        else:
            break

    # Release all space and windows once done
    cam.release()
    cv2.destroyAllWindows()
```

```python
cam = cv2.VideoCapture("S:/selfdrivingcar/videos/output.ogv")
create_frames(cam, 0)
```

```python
lst = os.listdir(directory)
cam2 = cv2.VideoCapture("S:/selfdrivingcar/videos/output4.ogv")
create_frames(cam2, len(lst))
```
