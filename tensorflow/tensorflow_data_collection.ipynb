# Set Paths and Create Folders

```python
VIDEO_PATH = 'S:/selfdrivingcar/videos'    
IMAGE_PATH = "S:/selfdrivingcar/data/images"
LABEL_PATH = "S:/selfdrivingcar/data/intersection"
LANE_PATH = "S:/selfdrivingcar/data/lane"

if not os.path.exists(IMAGE_PATH):
    os.makedirs(IMAGE_PATH)
if not os.path.exists(LABEL_PATH):
    os.makedirs(LABEL_PATH)
```

# Define Function for Frame Extraction

```python
def create_frames(cam, photo_id):
    currentframe = 0
    while(True):
        # reading from frame
        ret,frame = cam.read()

        if ret:
            # if video is still left continue creating images
            name = os.path.join(IMAGE_PATH, "frame_" + str(currentframe + photo_id) + ".jpg")
            print ('Creating...' + name)

            # writing the extracted images
            cv2.imwrite(name, frame)

            # increasing counter so that it will
            # show how many frames are created
            currentframe += 1
        else:
            break

    # Release all space and windows once done
    cam.release()
    cv2.destroyAllWindows()
```

# Extract Frames using OpenCV

```python
cam = cv2.VideoCapture("S:/selfdrivingcar/videos/output.ogv")
create_frames(cam, 0)

lst = os.listdir(IMAGE_PATH)
cam2 = cv2.VideoCapture("S:/selfdrivingcar/videos/output4.ogv")
create_frames(cam2, len(lst))

lst = os.listdir(IMAGE_PATH)
print(len(lst))
```
